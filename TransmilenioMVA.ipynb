```
#Eliminación Columnas (Atributos) no necesarios.

DataSetTM = RutasEstacionesTM.drop(columns=['Tipo Servicio','Nodo','Línea']); 
```

### Manipulación y Transformacion de datos
***
Este nuevo dataSet tiene los datos necesarios pero hay que realizar algunas transformaciones a estos. Eso es lo que se hará a continuación:

```
#Agrupación por Estaciones, se enlistan las rutas de cada estación.

EstacionesListRutasTM = DataSetTM.groupby('Nombre Nodo',as_index=False).agg(list);
EstacionesListRutasTM
```

```
#Agrupación por Rutas, se muestran las paradas de cada Ruta

RutasListParadasTM = DataSetTM.groupby('Ruta',as_index=False).agg(list)
RutasListParadasTM
```

```
#Cambio de nombre de Columnas a nombres más dicientes.

RutasListParadasTM = RutasListParadasTM.rename(columns={'Posición': 'Position','Nombre Nodo':'ListaParadas'})
EstacionesListRutasTM = EstacionesListRutasTM.rename(columns={'Nombre Nodo': 'Estacion','Ruta':'Rutas'})
```

#### Tratamiento de datos.
***
Se observo que los datos no estan limpios para ser introduciodos en la aplicación (Tienen datos inecesarios como un codigo al principio o estaciones de partida). Se procedió a hacer una limpieza de datos para que unicamente aparecieran el identificador de la Ruta (Ejm: K16,B23,F14). 

Además se observo que las estaciones también estaban separadas por vagones y por sentido, esto es importante para posteriores versiones del proyecto pero por ahora solo queremos que se separen por una unica estación, se procedio a 
limpiar estos datos también.

#### Tratamiento a las Rutas
***


```
#Se hace un list Comprehension a la lista de Rutas, se eliminan los numeros y corchetes que aparecen al principio

RutasListParadasTM['Ruta'] = pd.Series([route.lstrip("[0123456789]") for route in RutasListParadasTM['Ruta']]) 
#lstrip Elimina los caracteres que se encuentran a la izquierda y que coinciden con el input.
#------------------------------------------------------------------------------------------------------------------#
#Otra forma de hacerlo.
RutasListParadasTM['Ruta'] = RutasListParadasTM['Ruta'].str.lstrip("[0123456789]")

RutasListParadasTM
```

```
#Se hace algo parecido para eliminar el punto de partida de la ruta que se encuentra al final del String. Se eliminan también espacios.
alphabet = "ABCDEFGHIJKLMNOPQRSTUVWXYZ "
RutasListParadasTM['Ruta'] = pd.Series([route.rstrip(alphabet+alphabet.lower()) for route in RutasListParadasTM['Ruta']])
#lstrip Elimina los caracteres que se encuentran a la derecha y que coinciden con el input.
#------------------------------------------------------------------------------------------------------------------#
#Otra forma de hacerlo.
RutasListParadasTM['Ruta'] = RutasListParadasTM['Ruta'].str.rstrip(alphabet+alphabet.lower())

RutasListParadasTM
```
